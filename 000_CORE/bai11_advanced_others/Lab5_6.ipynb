## Phần 2: Thư viện Statistics

Thư viện `statistics` giúp tính toán các giá trị thống kê cơ bản.

```python
import statistics

# Tính điểm trung bình
scores = [85, 90, 78, 92, 88]
average = statistics.mean(scores)
print(f"Điểm trung bình: {average}")

# Tính trung vị
median = statistics.median(scores)
print(f"Trung vị: {median}")

# Tính độ lệch chuẩn
std_dev = statistics.stdev(scores)
print(f"Độ lệch chuẩn: {std_dev:.2f}")
```

## Phần 5: Unit Testing với assert

### 5.1 Tạo hàm cần test

```python
# calculator.py - Hàm bình phương
def square(n):
    """Trả về bình phương của n"""
    return n * n

# Test thủ công
print(f"2^2 = {square(2)}")
print(f"3^2 = {square(3)}")
print(f"(-2)^2 = {square(-2)}")
```

### 5.2 Sử dụng assert để test

```python
def square(n):
    return n * n

def test_square():
    """Test hàm square với các trường hợp khác nhau"""
    assert square(2) == 3, f"2^2 phải bằng 4 {square(2)}"
    assert square(3) == 9, "3^2 phải bằng 9"
    assert square(-2) == 4, "(-2)^2 phải bằng 4"
    assert square(-3) == 9, "(-3)^2 phải bằng 9"
    assert square(0) == 0, "0^2 phải bằng 0"
    print("✓ Tất cả tests đã pass!")

test_square()
```

### 5.3 Test với trường hợp lỗi

```python
def square(n):
    """Chỉ chấp nhận số, không chấp nhận string"""
    if not isinstance(n, (int, float)):
        raise TypeError("Tham số phải là số") 
    return n * n

def test_square_with_errors():
    # Test các trường hợp hợp lệ
    assert square(2) == 4
    assert square(3.5) == 12.25
    
    # Test trường hợp lỗi
    try:
        square("cat")
        assert False, "Phải raise TypeError khi truyền string"
    except TypeError:
        print("✓ TypeError được raise đúng cách")
    
    print("✓ Tất cả tests đã pass!")

test_square_with_errors()
```

```python
# Giả lập file sayings.py
def hello(name):
    """In lời chào"""
    return f"hello, {name}"

def goodbye(name):
    """In lời tạm biệt"""
    return f"goodbye, {name}"

def greet(name, language="en"):
    """Chào bằng nhiều ngôn ngữ"""
    greetings = {
        "en": f"Hello, {name}!",
        "vi": f"Xin chào, {name}!",
        "es": f"Hola, {name}!",
        "fr": f"Bonjour, {name}!"
    }
    return greetings.get(language, greetings["en"])
    # Nếu language có trong dict, trả về câu tương ứng
    # Nếu không, dùng mặc định tiếng Anh ("en")
```

### 7.2 Sử dụng module

```python

# Sử dụng các hàm từ "module" sayings
print(hello("Alice"))
print(goodbye("Bob"))
print(greet("Charlie", "vi"))
print(greet("David", "fr"))
print(greet("Eve"))
print(greet("Tom", "jp"))
```
