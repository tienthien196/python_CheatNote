```python
#!/usr/bin/env python3
"""
=============================================================================
PYTHON DATA TYPES CHEAT SHEET - INT, FLOAT, COMPLEX
=============================================================================
"""

# =============================================================================
# 1. INTEGER - C√ÅC H·ªÜ S·ªê
# =============================================================================

# --- Binary (Nh·ªã ph√¢n) - prefix: 0b ho·∫∑c 0B ---
0b0101      # 5
0B1111      # 15

# --- Octal (B√°t ph√¢n) - prefix: 0o ho·∫∑c 0O ---
0o7777      # 4095
0O4567      # 2423

# --- Hexadecimal (Th·∫≠p l·ª•c ph√¢n) - prefix: 0x ho·∫∑c 0X ---
0x0000      # 0
0XFFFF      # 65535

# --- Decimal (Th·∫≠p ph√¢n) - H·ªá c∆° s·ªë 10 m·∫∑c ƒë·ªãnh ---
255         # 255


# =============================================================================
# 2. CHUY·ªÇN ƒê·ªîI H·ªÜ S·ªê
# =============================================================================

x = 255
bin(x)      # '0b11111111' - chuy·ªÉn sang binary string
oct(x)      # '0o377' - chuy·ªÉn sang octal string
hex(x)      # '0xff' - chuy·ªÉn sang hexadecimal string

# Chuy·ªÉn ng∆∞·ª£c t·ª´ string v·ªÅ int
int('0b11111111', 2)    # 255 - t·ª´ binary
int('0o377', 8)         # 255 - t·ª´ octal
int('0xff', 16)         # 255 - t·ª´ hex
int('255', 10)          # 255 - t·ª´ decimal


# =============================================================================
# 3. UNDERSCORE TRONG S·ªê - Format hi·ªÉn th·ªã ƒë·∫πp
# =============================================================================

# Underscore (_) gi√∫p d·ªÖ ƒë·ªçc, kh√¥ng ·∫£nh h∆∞·ªüng gi√° tr·ªã
1_000_000           # 1000000
1_00_0000           # 1000000 - v·ªã tr√≠ underscore t√πy √Ω

# D√πng v·ªõi m·ªçi h·ªá s·ªë
0b01_10             # 6
0o77_77             # 4095
0xFF_FF_FF          # 16777215

# D√πng v·ªõi float
1_000.567_89        # 1000.56789


# =============================================================================
# 4. TYPE ANNOTATION - Ch√∫ th√≠ch ki·ªÉu
# =============================================================================

# ‚ö†Ô∏è Type annotation KH√îNG ·∫£nh h∆∞·ªüng runtime!
# Ch·ªâ l√† g·ª£i √Ω cho developer v√† type checker (mypy)

i: int = 34.6798        # V·∫´n l√† float! Type hint kh√¥ng b·∫Øt bu·ªôc
x: int = 100            # OK
y: float = 3.14         # OK

# ƒê·ªÉ b·∫Øt l·ªói type ‚Üí d√πng mypy:
# $ mypy script.py


# =============================================================================
# 5. FLOAT - S·ªë th·ª±c
# =============================================================================

# --- Float c∆° b·∫£n ---
f: float = 3.14
1.5
0.5
.5              # C√≥ th·ªÉ b·ªè 0 ƒë·∫ßu

# --- Scientific notation (K√Ω hi·ªáu khoa h·ªçc) ---
# e ho·∫∑c E ƒë·ªÅu t·∫°o float (exponent - s·ªë m≈©)
1e2             # 100.0 = 1 √ó 10¬≤
1E5             # 100000.0 = 1 √ó 10‚Åµ
2e3             # 2000.0 = 2 √ó 10¬≥
2.5e-3          # 0.0025 = 2.5 √ó 10‚Åª¬≥
2.5E-3          # 0.0025 = 2.5 √ó 10‚Åª¬≥

# ‚ö†Ô∏è L∆∞u √Ω: e/E lu√¥n t·∫°o float, KH√îNG ph·∫£i int
1e2             # 100.0 (float)
type(1e2)       # <class 'float'>

# --- Float v·ªõi underscore ---
1_000.5         # 1000.5
3.141_592_653   # 3.141592653


# =============================================================================
# 5A. FLOAT FORMATTING - f-string format specifier
# C√∫ ph√°p chu·∫©n: f"{value:[fill][align][sign][#][0][width][,][.precision][type]}"
# =============================================================================

# üî¢ 1. Gi·ªõi h·∫°n s·ªë ch·ªØ s·ªë th·∫≠p ph√¢n
x = 3.14159265
f"{x:.2f}"      # ‚Üí "3.14"
f"{x:.4f}"      # ‚Üí "3.1416" (l√†m tr√≤n)
f"{x:.6f}"      # ‚Üí "3.141593"

# üìè 2. CƒÉn l·ªÅ & ƒë·ªô r·ªông tr∆∞·ªùng (width & alignment)
x = 12.34
f"{x:10.2f}"    # ‚Üí "     12.34" (t·ªïng 10 k√Ω t·ª±, cƒÉn ph·∫£i m·∫∑c ƒë·ªãnh)
f"{x:>10.2f}"   # ‚Üí "     12.34" (cƒÉn ph·∫£i - explicit)
f"{x:<10.2f}"   # ‚Üí "12.34     " (cƒÉn tr√°i)
f"{x:^10.2f}"   # ‚Üí "  12.34   " (cƒÉn gi·ªØa)

# üîÅ 3. D·∫•u ph√¢n c√°ch h√†ng ng√†n
x = 1234567.89
f"{x:,.2f}"     # ‚Üí "1,234,567.89" (d·∫•u ph·∫©y)
f"{x:_.2f}"     # ‚Üí "1_234_567.89" (underscore)

# üß™ 4. K√Ω hi·ªáu khoa h·ªçc (scientific notation)
x = 0.000123456
f"{x:.2e}"      # ‚Üí "1.23e-04" (ch·ªØ e th∆∞·ªùng)
f"{x:.2E}"      # ‚Üí "1.23E-04" (ch·ªØ E hoa)

# üîÑ 5. T·ª± ƒë·ªông ch·ªçn gi·ªØa f v√† e (d√πng g ho·∫∑c G)
x = 0.000123456
f"{x:.2g}"      # ‚Üí "0.00012" (ch·ªçn format ng·∫Øn nh·∫•t)
x = 1234567.89
f"{x:.2g}"      # ‚Üí "1.2e+06" (t·ª± chuy·ªÉn sang e khi s·ªë l·ªõn)

# üìä 6. Hi·ªÉn th·ªã d∆∞·ªõi d·∫°ng ph·∫ßn trƒÉm (%)
x = 0.857
f"{x:.1%}"      # ‚Üí "85.7%"
f"{x:.2%}"      # ‚Üí "85.70%"

# üî≤ 7. ƒêi·ªÅn s·ªë 0 ·ªü ƒë·∫ßu (zero-padding)
x = 42.5
f"{x:08.2f}"    # ‚Üí "00042.50" (t·ªïng 8 k√Ω t·ª±, pad b·∫±ng 0)
f"{x:010.3f}"   # ‚Üí "000042.500"

# üßÆ 8. Hi·ªÉn th·ªã d·∫•u (+ / -) r√µ r√†ng
x = -3.14
y = 2.71
f"{x:+.2f}"     # ‚Üí "-3.14"
f"{y:+.2f}"     # ‚Üí "+2.71"
f"{y: .2f}"     # ‚Üí " 2.71" (d·∫•u c√°ch thay cho +)

# üéØ 9. K·∫øt h·ª£p nhi·ªÅu format
x = -1234.5678
f"{x:+012,.2f}" # ‚Üí "-01,234.57" (d·∫•u +, pad 0, ph·∫©y, 2 ch·ªØ s·ªë th·∫≠p ph√¢n)
f"{x:+015_.3f}" # ‚Üí "-00001_234.568" (underscore separator)

# üìê 10. Format v·ªõi fill character t√πy ch·ªânh
x = 42.5
f"{x:*>10.2f}"  # ‚Üí "****42.50" (fill b·∫±ng *)
f"{x:_^12.2f}"  # ‚Üí "___42.50____" (cƒÉn gi·ªØa, fill b·∫±ng _)


# =============================================================================
# 5B. FLOATING-POINT PRECISION ISSUES - V·∫•n ƒë·ªÅ ƒë·ªô ch√≠nh x√°c
# =============================================================================

# ‚ö†Ô∏è IEEE 754 representation error
# Float ƒë∆∞·ª£c l∆∞u d∆∞·ªõi d·∫°ng binary ‚Üí kh√¥ng bi·ªÉu di·ªÖn ch√≠nh x√°c decimal

# Binary floating-point inaccuracy
0.1 + 0.2               # ‚Üí 0.30000000000000004 (KH√îNG ph·∫£i 0.3!)
0.1 + 0.2 == 0.3        # ‚Üí False (!)

1.2 + 0.3               # ‚Üí 1.5 (ƒë√∫ng - may m·∫Øn)
1.2 + 0.3 == 1.5        # ‚Üí True

# Decimal fraction representation problem
0.1                     # ‚Üí 0.1000000000000000055511151231257827...
0.2                     # ‚Üí 0.2000000000000000111022302462515654...
0.3                     # ‚Üí 0.2999999999999999888977697537484345...

# ‚úÖ Gi·∫£i ph√°p 1: So s√°nh v·ªõi epsilon (ƒë·ªô sai s·ªë nh·ªè)
import math
abs(0.1 + 0.2 - 0.3) < 1e-9         # ‚Üí True
math.isclose(0.1 + 0.2, 0.3)        # ‚Üí True (Python 3.5+)

# ‚úÖ Gi·∫£i ph√°p 2: D√πng decimal module cho ƒë·ªô ch√≠nh x√°c cao
from decimal import Decimal
Decimal('0.1') + Decimal('0.2')     # ‚Üí Decimal('0.3')
Decimal('0.1') + Decimal('0.2') == Decimal('0.3')  # ‚Üí True

# ‚úÖ Gi·∫£i ph√°p 3: L√†m tr√≤n khi so s√°nh
round(0.1 + 0.2, 10) == round(0.3, 10)  # ‚Üí True


# =============================================================================
# 6. COMPLEX - S·ªë ph·ª©c
# =============================================================================

# D·∫°ng: a + bj (trong ƒë√≥ j = ‚àö-1)
3+4j            # (3+4j)
3+4J            # (3+4j) - J ho·∫∑c j ƒë·ªÅu ƒë∆∞·ª£c
5j              # 5j = 0+5j
-2-3j           # (-2-3j)

# V·ªõi type annotation
v: complex = 3+4j

# C√°c thu·ªôc t√≠nh
z = 3+4j
z.real          # 3.0 - ph·∫ßn th·ª±c
z.imag          # 4.0 - ph·∫ßn ·∫£o
abs(z)          # 5.0 - module |z| = ‚àö(3¬≤+4¬≤)
z.conjugate()   # (3-4j) - s·ªë ph·ª©c li√™n h·ª£p

# H√†m complex()
complex(3, 4)   # (3+4j)
complex(5)      # (5+0j)
complex('3+4j') # (3+4j)

# ‚ö†Ô∏è L∆ØU √ù: Complex KH√îNG so s√°nh ƒë∆∞·ª£c th·ª© t·ª±
# 3+4j > 2+1j   # TypeError: '>' not supported
# 3+4j < 2+1j   # TypeError: '<' not supported

# Ch·ªâ so s√°nh == v√† !=
3+4j == 3+4j    # True
3+4j != 2+1j    # True


# =============================================================================
# 7. PH√âP TO√ÅN V√Ä CHIA
# =============================================================================

# Division (/) lu√¥n tr·∫£ v·ªÅ float
10 / 2          # 5.0 (float)
10 / 3          # 3.3333333333333335 (float)

# Integer division (//) - chia l·∫•y ph·∫ßn nguy√™n
10 // 3         # 3 (int)
10 // 2         # 5 (int)
-10 // 3        # -4 (l√†m tr√≤n xu·ªëng)

# Modulo (%) - chia l·∫•y ph·∫ßn d∆∞
10 % 3          # 1
10 % 2          # 0

# Power (**) - l≈©y th·ª´a
2 ** 3          # 8
2 ** 0.5        # 1.4142135623730951 (cƒÉn b·∫≠c 2)


# =============================================================================
# 8. KI·ªÇM TRA KI·ªÇU D·ªÆ LI·ªÜU
# =============================================================================

type(100)       # <class 'int'>
type(3.14)      # <class 'float'>
type(3+4j)      # <class 'complex'>
type(1e2)       # <class 'float'>

isinstance(100, int)        # True
isinstance(3.14, float)     # True
isinstance(3+4j, complex)   # True
isinstance(1e2, float)      # True


# =============================================================================
# 9. CHUY·ªÇN ƒê·ªîI KI·ªÇU
# =============================================================================

# Sang int
int(3.14)       # 3 - c·∫Øt b·ªè ph·∫ßn th·∫≠p ph√¢n
int('100')      # 100
int(True)       # 1
int(False)      # 0

# Sang float
float(100)      # 100.0
float('3.14')   # 3.14
float('1e2')    # 100.0

# Sang complex
complex(5)      # (5+0j)
complex(3, 4)   # (3+4j)


# =============================================================================
# 10. ƒê·∫∂C BI·ªÜT V√Ä L∆ØU √ù
# =============================================================================

# Infinity
float('inf')    # inf
float('-inf')   # -inf

# NaN (Not a Number)
float('nan')    # nan

# Bool l√† subclass c·ªßa int
True + True     # 2
True * 5        # 5
False + 10      # 10

# Division by zero
# 1 / 0         # ZeroDivisionError
# 1 // 0        # ZeroDivisionError
# 1.0 / 0         # ZeroDivisionError (v·∫´n l·ªói v·ªõi float)


# =============================================================================
# T·ªîNG K·∫æT
# =============================================================================

# ‚úì Underscore (_) trong s·ªë: d·ªÖ ƒë·ªçc, kh√¥ng ·∫£nh h∆∞·ªüng gi√° tr·ªã
# ‚úì Type annotation: ch·ªâ g·ª£i √Ω, kh√¥ng b·∫Øt bu·ªôc runtime
# ‚úì e/E: lu√¥n t·∫°o float (scientific notation)
# ‚úì Division (/): lu√¥n tr·∫£ float
# ‚úì Integer division (//): tr·∫£ int
# ‚úì Complex: kh√¥ng so s√°nh ƒë∆∞·ª£c >, <, >=, <=
# ‚úì Bool l√† subclass c·ªßa int
```
