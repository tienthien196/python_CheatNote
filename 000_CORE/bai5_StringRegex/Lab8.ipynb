## Ph·∫ßn 3: Case Sensitivity - X·ª≠ l√Ω ch·ªØ hoa/th∆∞·ªùng

### S·ª≠ d·ª•ng flag re.IGNORECASE

```python
import re

email = input("What's your email? ").strip()

# re.IGNORECASE: kh√¥ng ph√¢n bi·ªát hoa th∆∞·ªùng
if re.search(r"^\w+@\w+\.edu$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")

# Gi·ªù "MALAN@HARVARD.EDU" c≈©ng h·ª£p l·ªá
```

### X·ª≠ l√Ω subdomain (v√≠ d·ª•: malan@cs50.harvard.edu)

```python
import re

email = input("What's your email? ").strip()

# (\w+\.)? nghƒ©a l√† "c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng c√≥ subdomain"
# ?: trong (?:...) nghƒ©a l√† "non-capturing group" - kh√¥ng l∆∞u k·∫øt qu·∫£
if re.search(r"^\w+@(\w+\.)?\w+\.edu$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")

# C·∫£ "malan@harvard.edu" v√† "malan@cs50.harvard.edu" ƒë·ªÅu h·ª£p l·ªá
```

### S·ª≠ d·ª•ng regex ƒë·ªÉ x·ª≠ l√Ω linh ho·∫°t h∆°n

```python
import re

name = input("What's your name? ").strip()

# ^(.+), *(.+)$ nghƒ©a l√†:
# - B·∫Øt ƒë·∫ßu v·ªõi m·ªôt ho·∫∑c nhi·ªÅu k√Ω t·ª± (group 1)
# - Theo sau l√† d·∫•u ph·∫©y v√† kh√¥ng ho·∫∑c nhi·ªÅu kho·∫£ng tr·∫Øng
# - K·∫øt th√∫c v·ªõi m·ªôt ho·∫∑c nhi·ªÅu k√Ω t·ª± (group 2)
matches = re.search(r"^(.+), *(.+)$", name)

if matches:
    last, first = matches.groups()
    name = f"{first} {last}"

print(f"hello, {name}")
```

### S·ª≠ d·ª•ng group() ƒë·ªÉ truy xu·∫•t t·ª´ng nh√≥m

```python
import re

name = input("What's your name? ").strip()

matches = re.search(r"^(.+), *(.+)$", name)

if matches:
    # group(1) l√† nh√≥m ƒë·∫ßu ti√™n (h·ªç), group(2) l√† nh√≥m th·ª© hai (t√™n)
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

### S·ª≠ d·ª•ng walrus operator :=

```python
import re

name = input("What's your name? ").strip()

# := g√°n gi√° tr·ªã V√Ä ki·ªÉm tra ƒëi·ªÅu ki·ªán c√πng l√∫c
if matches := re.search(r"^(.+), *(.+)$", name):
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

### X·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p linh ho·∫°t h∆°n

```python
import re

url = input("URL: ").strip()

# ^(https?://)? - c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng http ho·∫∑c https
# (www\.)? - c√≥ th·ªÉ c√≥ ho·∫∑c kh√¥ng www.
# twitter\.com/ - b·∫Øt bu·ªôc c√≥ twitter.com/
username = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url)
print(f"Username: {username}")

# Gi·ªù c√≥ th·ªÉ x·ª≠ l√Ω:
# - https://twitter.com/davidjmalan
# - http://twitter.com/davidjmalan
# - www.twitter.com/davidjmalan
# - twitter.com/davidjmalan
```

### S·ª≠ d·ª•ng re.search() ƒë·ªÉ validate v√† extract

```python
import re

url = input("URL: ").strip()

# (.+)$ capture username ·ªü cu·ªëi URL
matches = re.search(r"^https?://(www\.)?twitter\.com/(.+)$", url, re.IGNORECASE)

if matches:
    print(f"Username: {matches.group(2)}")
else:
    print("Invalid URL")
```

### S·ª≠ d·ª•ng non-capturing group v√† walrus operator

```python
import re

url = input("URL: ").strip()

# (?:www\.)? - non-capturing group, kh√¥ng ƒë·∫øm v√†o s·ªë group
if matches := re.search(r"^https?://(?:www\.)?twitter\.com/(.+)$", url, re.IGNORECASE):
    print(f"Username: {matches.group(1)}")  # Gi·ªù username l√† group 1 thay v√¨ 2
else:
    print("Invalid URL")
```

### Validate username theo quy t·∫Øc c·ªßa Twitter

```python
import re

url = input("URL: ").strip()

# [a-z0-9_]+ - username ch·ªâ ch·ª©a ch·ªØ, s·ªë, g·∫°ch d∆∞·ªõi
if matches := re.search(r"^https?://(?:www\.)?twitter\.com/([a-z0-9_]+)$", url, re.IGNORECASE):
    print(f"Username: {matches.group(1)}")
else:
    print("Invalid URL")
```

### Regex Cheat Sheet

#### Anchors (Neo)
- `^` - B·∫Øt ƒë·∫ßu chu·ªói/d√≤ng
- `$` - K·∫øt th√∫c chu·ªói/d√≤ng
- `\b` - Word boundary
- `\B` - Not word boundary

#### Quantifiers (B·ªô ƒë·ªãnh l∆∞·ª£ng)
- `.` - 1
- `*` - 0 ho·∫∑c nhi·ªÅu
- `+` - 1 ho·∫∑c nhi·ªÅu
- `?` - 0 ho·∫∑c 1
- `{n}` - Ch√≠nh x√°c n l·∫ßn
- `{n,}` - n ho·∫∑c nhi·ªÅu l·∫ßn
- `{n,m}` - T·ª´ n ƒë·∫øn m l·∫ßn

#### Character Classes
- `\d` - Ch·ªØ s·ªë [0-9]
- `\D` - Kh√¥ng ph·∫£i ch·ªØ s·ªë
- `\w` - Word character [a-zA-Z0-9_]
- `\W` - Kh√¥ng ph·∫£i word character
- `\s` - Whitespace
- `\S` - Kh√¥ng ph·∫£i whitespace

#### Groups v√† Lookaround
- `(...)` - Capturing group
- `(?:...)` - Non-capturing group
- `(?=...)` - Positive lookahead
- `(?!...)` - Negative lookahead
