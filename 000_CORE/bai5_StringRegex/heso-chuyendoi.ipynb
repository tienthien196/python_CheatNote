```python
# int() với base - Chuyển đổi hệ số
# Chuyển chuỗi sang số với hệ cơ số khác nhau.

# Binary (base 2)
int("1010", 2)                  # 10
int("11111111", 2)              # 255

# Octal (base 8)
int("17", 8)                    # 15
int("377", 8)                   # 255

# Decimal (base 10) - mặc định
int("42")                       # 42
int("255", 10)                  # 255

# Hexadecimal (base 16)
int("ff", 16)                   # 255
int("FF", 16)                   # 255
int("10", 16)                   # 16
int("0xff", 16)                 # 255

# Base bất kỳ (2-36)
int("z", 36)                    # 35
int("10", 36)                   # 36

# Chuyển đổi ngược lại:

num = 255

# Sang binary
bin(num)                        # "0b11111111"
format(num, 'b')                # "11111111" (không có 0b)

# Sang octal
oct(num)                        # "0o377"
format(num, 'o')                # "377"

# Sang hex
hex(num)                        # "0xff"
format(num, 'x')                # "ff" (lowercase)
format(num, 'X')                # "FF" (uppercase)
```

```python
# Tổng kết các chuyển đổi:
# String ↔ Bytes
"hello".encode("utf-8")         # → b'hello'
b"hello".decode("utf-8")        # → "hello"

# Char ↔ Unicode code
ord("A")                        # → 65
chr(65)                         # → "A"

# String ↔ Number (different bases)
int("ff", 16)                   # → 255
hex(255)                        # → "0xff"
bin(255)                        # → "0b11111111"
oct(255)                        # → "0o377"
```
