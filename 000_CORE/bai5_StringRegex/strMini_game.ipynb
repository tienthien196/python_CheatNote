```python
# swapcase() - ƒê·ªïi hoa ‚Üî th∆∞·ªùng
# ƒê·ªïi ch·ªØ HOA ‚Üí th∆∞·ªùng v√† th∆∞·ªùng ‚Üí HOA.

s = "Hello World"
s.swapcase()                    # "hELLO wORLD"

"Python".swapcase()             # "pYTHON"
"ABC123xyz".swapcase()          # "abc123XYZ"
"HeLLo WoRLd".swapcase()        # "hEllO wOrlD"

# Unicode
"Caf√©".swapcase()               # "cAF√â"
# Use case (hi·∫øm khi d√πng):
# Toggle case
text = "Hello"
text = text.swapcase()          # "hELLO"
text = text.swapcase()          # "Hello" (quay l·∫°i)

# T·∫°o hi·ªáu ·ª©ng text
def alternating_case(s):
    result = ""
    for i, char in enumerate(s):
        if i % 2 == 0:
            result += char.upper()
        else:
            result += char.lower()
    return result

alternating_case("hello")       # "HeLlO"
```

```python
# casefold() - Chuy·ªÉn th√†nh ch·ªØ th∆∞·ªùng m·∫°nh (Unicode)
# Gi·ªëng lower() nh∆∞ng m·∫°nh h∆°n cho vi·ªác so s√°nh kh√¥ng ph√¢n bi·ªát hoa/th∆∞·ªùng v·ªõi Unicode.
s = "Hello"
s.casefold()                    # "hello" (gi·ªëng lower())

# S·ª± kh√°c bi·ªát v·ªõi lower()
"HELLO".lower()                 # "hello"
"HELLO".casefold()              # "hello" (gi·ªëng nhau)

# ƒêi·ªÉm kh√°c bi·ªát: Unicode ƒë·∫∑c bi·ªát
"Stra√üe".lower()                # "stra√üe" (gi·ªØ nguy√™n √ü)
"Stra√üe".casefold()             # "strasse" (√ü ‚Üí ss) ‚úÖ

# Ti·∫øng Hy L·∫°p
"Œ£".lower()                     # "œÉ"
"Œ£".casefold()                  # "œÉ" (gi·ªëng nhau trong tr∆∞·ªùng h·ª£p n√†y)

# Ti·∫øng Th·ªï Nhƒ© K·ª≥
"ƒ∞".lower()                     # "iÃá" (c√≥ d·∫•u ch·∫•m)
"ƒ∞".casefold()                  # "iÃá"
```

```python
# Khi n√†o d√πng casefold() thay v√¨ lower()?
# So s√°nh string kh√¥ng ph√¢n bi·ªát hoa/th∆∞·ªùng (Unicode)
def case_insensitive_equal(s1, s2):
    return s1.casefold() == s2.casefold()

# V√≠ d·ª• v·ªõi ti·∫øng ƒê·ª©c
street1 = "Stra√üe"
street2 = "STRASSE"

# D√πng lower() - SAI ‚ùå
street1.lower() == street2.lower()  # False ("stra√üe" != "strasse")

# D√πng casefold() - ƒê√öNG ‚úÖ
street1.casefold() == street2.casefold()  # True
# Use case:
# So s√°nh kh√¥ng ph√¢n bi·ªát hoa/th∆∞·ªùng (ƒë√∫ng chu·∫©n)
def search(query, text):
    return query.casefold() in text.casefold()

search("stra√üe", "Die Stra√üe ist lang")  # True

# T√¨m ki·∫øm trong list
words = ["Stra√üe", "Caf√©", "Python"]
query = "strasse"
found = [w for w in words if w.casefold() == query.casefold()]
# ['Stra√üe']

# Sort kh√¥ng ph√¢n bi·ªát hoa/th∆∞·ªùng
names = ["alice", "Bob", "CHARLIE"]
sorted(names, key=str.casefold)  # ['alice', 'Bob', 'CHARLIE']
```

```python
# # 8. CƒÇN CH·ªàNH V√Ä ƒê·ªäNH D·∫†NG
# center() - CƒÉn gi·ªØa
# CƒÉn gi·ªØa chu·ªói trong ƒë·ªô r·ªông cho tr∆∞·ªõc, ƒë·ªám k√Ω t·ª± fillchar hai b√™n.
s = "Hello"

# CƒÉn gi·ªØa v·ªõi kho·∫£ng tr·∫Øng (m·∫∑c ƒë·ªãnh)
s.center(11)                    # "   Hello   "
s.center(10)                    # "  Hello   " (l·∫ª th√¨ b√™n ph·∫£i nhi·ªÅu h∆°n)

# CƒÉn gi·ªØa v·ªõi k√Ω t·ª± t√πy ch·ªânh
s.center(11, "-")               # "---Hello---"
s.center(11, "*")               # "***Hello***"
s.center(15, "=")               # "=====Hello====="

# N·∫øu width <= len(string): gi·ªØ nguy√™n
s.center(3)                     # "Hello" (kh√¥ng c·∫Øt)
s.center(5)                     # "Hello"


# Use case:
# T·∫°o header
title = "MENU"
print(title.center(50, "="))
# ======================MENU=======================

# T·∫°o b·∫£ng ƒë·∫πp
print("Name".center(20) + "Age".center(10))
print("Alice".center(20) + "25".center(10))
```

```python
# ljust() - CƒÉn tr√°i (Left Justify)
# CƒÉn tr√°i chu·ªói, ƒë·ªám k√Ω t·ª± fillchar b√™n ph·∫£i.
s = "Hello"

# CƒÉn tr√°i v·ªõi kho·∫£ng tr·∫Øng
s.ljust(10)                     # "Hello     "
s.ljust(15)                     # "Hello          "

# CƒÉn tr√°i v·ªõi k√Ω t·ª± t√πy ch·ªânh
s.ljust(10, ".")                # "Hello....."
s.ljust(15, "-")                # "Hello----------"

# N·∫øu width <= len(string): gi·ªØ nguy√™n
s.ljust(3)                      # "Hello"

# Use case:

# Format b·∫£ng
items = [("Apple", 10), ("Banana", 5), ("Orange", 8)]
for name, price in items:
    print(name.ljust(15, ".") + str(price).rjust(5))
# Apple..........   10
# Banana.........    5
# Orange.........    8
```

```python
# rjust() - CƒÉn ph·∫£i (Right Justify)
# CƒÉn ph·∫£i chu·ªói, ƒë·ªám k√Ω t·ª± fillchar b√™n tr√°i.
s = "Hello"

# CƒÉn ph·∫£i v·ªõi kho·∫£ng tr·∫Øng
s.rjust(10)                     # "     Hello"
s.rjust(15)                     # "          Hello"

# CƒÉn ph·∫£i v·ªõi k√Ω t·ª± t√πy ch·ªânh
s.rjust(10, ".")                # ".....Hello"
s.rjust(15, "0")                # "0000000000Hello"


# Use case:
# Format s·ªë ti·ªÅn
prices = [100, 1500, 25000]
for price in prices:
    print(str(price).rjust(10) + " VND")
#       100 VND
#      1500 VND
#     25000 VND

# Format s·ªë c√≥ ƒë·ªô r·ªông c·ªë ƒë·ªãnh
# for i in range(1, 101):
#     print(str(i).rjust(3, "0"))  # 001, 002, ..., 100
```

```python
# zfill() - ƒê·ªám s·ªë 0 b√™n tr√°i
# ƒê·ªám s·ªë 0 b√™n tr√°i ƒë·ªÉ ƒë·∫°t ƒë·ªô r·ªông cho tr∆∞·ªõc. ƒê·∫∑c bi·ªát x·ª≠ l√Ω d·∫•u +/-.
# ƒê·ªám s·ªë 0
"42".zfill(5)                   # "00042"
"123".zfill(8)                  # "00000123"

# X·ª≠ l√Ω d·∫•u √¢m/d∆∞∆°ng
"-42".zfill(5)                  # "-0042" (d·∫•u - ·ªü ƒë·∫ßu)
"+42".zfill(5)                  # "+0042" (d·∫•u + ·ªü ƒë·∫ßu)

# N·∫øu width <= len(string): gi·ªØ nguy√™n
"12345".zfill(3)                # "12345"

# So s√°nh v·ªõi rjust()
"42".rjust(5, "0")              # "00042" (gi·ªëng)
"-42".rjust(5, "0")             # "00-42" ‚ùå (sai v·ªã tr√≠ d·∫•u)
"-42".zfill(5)                  # "-0042" ‚úÖ (ƒë√∫ng)

# Use case:
# Format ID
user_id = 42
print(f"USER_{str(user_id).zfill(6)}")  # USER_000042

# ƒê√°nh s·ªë file
for i in range(1, 101):
    filename = f"image_{str(i).zfill(4)}.png"
    print(filename)  # image_0001.png, image_0002.png, ...

# Format m√£ ƒë∆°n h√†ng
order = 123
print(f"ORD-{str(order).zfill(8)}")  # ORD-00000123
```

```python
# format() method
# C√°ch c≈© h∆°n nh∆∞ng v·∫´n h·ªØu √≠ch.

# V·ªã tr√≠ (positional)
"{} is {}".format("Python", "awesome")  # "Python is awesome"
"{0} {1} {0}".format("A", "B")         # "A B A"

# T·ª´ kh√≥a (keyword)
"{lang} is {adj}".format(lang="Python", adj="great")
# "Python is great"

# Tr·ªôn l·∫´n
"{0} is {adj}".format("Python", adj="cool")
# "Python is cool"

# Format s·ªë
"{:.2f}".format(3.14159)        # "3.14"
"{:,}".format(1000000)          # "1,000,000"

# CƒÉn ch·ªânh
"{:>10}".format("Hi")           # "        Hi"
"{:<10}".format("Hi")           # "Hi        "
"{:^10}".format("Hi")           # "    Hi    "
"{:*^10}".format("Hi")          # "****Hi****"

# T·ª´ dictionary
data = {"name": "Alice", "age": 25}
"{name} is {age}".format(**data)  # "Alice is 25"
```

```python
# printf-style (%-formatting) - C≈®, KH√îNG KHUY√äN

# C∆° b·∫£n
"%s is %d years old" % ("Alice", 25)    # "Alice is 25 years old"

# Format specifiers
"%d" % 42                       # "42" (integer)
"%f" % 3.14                     # "3.140000" (float)
"%.2f" % 3.14159                # "3.14" (2 decimal places)
"%s" % "Hello"                  # "Hello" (string)

# Width v√† precision
"%10s" % "Hi"                   # "        Hi" (width 10)
"%-10s" % "Hi"                  # "Hi        " (left align)
"%10d" % 42                     # "        42"
"%010d" % 42                    # "0000000042" (zero padding)

# Nhi·ªÅu gi√° tr·ªã
"%s: %d/%d" % ("Score", 85, 100)  # "Score: 85/100"

# T·ª´ dictionary
"%(name)s is %(age)d" % {"name": "Alice", "age": 25}
# "Alice is 25"
```

```python
# 10. M√É H√ìA V√Ä CHUY·ªÇN ƒê·ªîI
# encode() - Chuy·ªÉn string ‚Üí bytes
# Chuy·ªÉn chu·ªói Unicode th√†nh bytes v·ªõi m√£ h√≥a ch·ªâ ƒë·ªãnh.
s = "Hello"

# M√£ h√≥a UTF-8 (m·∫∑c ƒë·ªãnh)
s.encode()                      # b'Hello'
s.encode("utf-8")               # b'Hello'

# Ti·∫øng Vi·ªát
"Xin ch√†o".encode("utf-8")      # b'Xin ch\xc3\xa0o'

# M√£ h√≥a kh√°c
"Hello".encode("ascii")         # b'Hello'
"Hello".encode("utf-16")        # b'\xff\xfeH\x00e\x00l\x00l\x00o\x00'

# X·ª≠ l√Ω l·ªói
"caf√©".encode("ascii")          # ‚ùå UnicodeEncodeError

# errors="ignore" - b·ªè qua k√Ω t·ª± l·ªói
"caf√©".encode("ascii", errors="ignore")     # b'caf'

# errors="replace" - thay b·∫±ng ?
"caf√©".encode("ascii", errors="replace")    # b'caf?'

# errors="xmlcharrefreplace" - thay b·∫±ng XML entity
"caf√©".encode("ascii", errors="xmlcharrefreplace")  # b'caf&#233;'
```

```python
# decode() - Chuy·ªÉn bytes ‚Üí string
# Chuy·ªÉn bytes th√†nh chu·ªói Unicode.

# Decode UTF-8
b"Hello".decode("utf-8")        # "Hello"
b"Xin ch\xc3\xa0o".decode("utf-8")  # "Xin ch√†o"

# Decode ASCII
b"Hello".decode("ascii")        # "Hello"

# X·ª≠ l√Ω l·ªói
b"\xff\xfe".decode("utf-8")     # ‚ùå UnicodeDecodeError

# errors="ignore" - b·ªè qua byte l·ªói
b"Hello\xff".decode("utf-8", errors="ignore")  # "Hello"

# errors="replace" - thay b·∫±ng ÔøΩ
b"Hello\xff".decode("utf-8", errors="replace")  # "HelloÔøΩ"
```

```python
# ·ª®ng d·ª•ng:
# Caesar cipher
def encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

encrypt("HELLO", 3)             # "KHOOR"

# T·∫°o b·∫£ng ASCII
for i in range(65, 91):
    print(f"{i}: {chr(i)}")     # 65: A, 66: B, ...
```
