## 4. ƒê·ªçc File

### 4.1. ƒê·ªçc to√†n b·ªô file v·ªõi `readlines()`

```python
# V√≠ d·ª• 4: ƒê·ªçc file names.txt
with open("names.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    print("hello,", line.rstrip())  # rstrip() lo·∫°i b·ªè \n ·ªü cu·ªëi
```

### 4.2. ƒê·ªçc t·ª´ng d√≤ng m·ªôt (hi·ªáu qu·∫£ h∆°n)

```python
# V√≠ d·ª• 5: ƒê·ªçc v√† s·∫Øp x·∫øp t√™n
names = []

with open("names.txt", "r") as file:
    for line in file:
        names.append(line.rstrip())

# S·∫Øp x·∫øp v√† in ra
for name in sorted(names):
    print(f"hello, {name}")
```

## 5. L√†m vi·ªác v·ªõi CSV Files

**CSV** (Comma Separated Values) l√† ƒë·ªãnh d·∫°ng file ph·ªï bi·∫øn ƒë·ªÉ l∆∞u d·ªØ li·ªáu d·∫°ng b·∫£ng.

### 5.1. T·∫°o file CSV m·∫´u

```python
# T·∫°o file students.csv
with open("students.csv", "w") as file:
    file.write("Hermione,Gryffindor\n")
    file.write("Harry,Gryffindor\n")
    file.write("Ron,Gryffindor\n")
    file.write("Draco,Slytherin\n")

print("ƒê√£ t·∫°o file students.csv")
```

### 5.2. ƒê·ªçc CSV kh√¥ng d√πng th∆∞ vi·ªán

```python
# V√≠ d·ª• 6: ƒê·ªçc CSV b·∫±ng split()
with open("students.csv") as file:
    for line in file:
        name, house = line.rstrip().split(",")
        print(f"{name} is in {house}")
```

### 5.3. L∆∞u v√†o dictionary v√† s·∫Øp x·∫øp

```python
# V√≠ d·ª• 7: S·ª≠ d·ª•ng dictionary
students = []

with open("students.csv") as file:
    for line in file:
        name, house = line.rstrip().split(",")
        students.append({"name": name, "house": house})

# S·∫Øp x·∫øp theo t√™n s·ª≠ d·ª•ng lambda function
for student in sorted(students, key=lambda student: student["name"]):
    print(f"{student['name']} is in {student['house']}")
```

## 6. S·ª≠ d·ª•ng th∆∞ vi·ªán `csv`

Python c√≥ th∆∞ vi·ªán `csv` t√≠ch h·ª£p s·∫µn ƒë·ªÉ x·ª≠ l√Ω CSV files hi·ªáu qu·∫£ h∆°n.

### 6.1. T·∫°o file CSV v·ªõi header

```python
# T·∫°o file v·ªõi header
with open("students.csv", "w") as file:
    file.write("name,home\n")
    file.write('Harry,"Number Four, Privet Drive"\n')
    file.write("Ron,The Burrow\n")
    file.write("Draco,Malfoy Manor\n")

print("ƒê√£ t·∫°o file students.csv v·ªõi header")
```

### 6.2. ƒê·ªçc CSV v·ªõi `csv.reader()`

```python
# V√≠ d·ª• 8: S·ª≠ d·ª•ng csv.reader
import csv

students = []

with open("students.csv") as file:
    reader = csv.reader(file)
    next(reader)  # B·ªè qua d√≤ng header
    
    for row in reader:
        students.append({"name": row[0], "home": row[1]})

for student in sorted(students, key=lambda s: s["name"]):
    print(f"{student['name']} is from {student['home']}")
```

### 6.3. ƒê·ªçc CSV v·ªõi `csv.DictReader()` (Khuy√™n d√πng)

```python
# V√≠ d·ª• 9: S·ª≠ d·ª•ng csv.DictReader - T·ª± ƒë·ªông x·ª≠ l√Ω header
import csv

students = []

with open("students.csv") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        students.append({"name": row["name"], "home": row["home"]})

for student in sorted(students, key=lambda s: s["name"]):
    print(f"{student['name']} is from {student['home']}")
```

### 6.4. Ghi CSV v·ªõi `csv.DictWriter()`

```python
# V√≠ d·ª• 10: Ghi d·ªØ li·ªáu v√†o CSV
import csv

name = input("What's your name? ")
home = input("Where's your home? ")

with open("students.csv", "a", newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["name", "home"])
    writer.writerow({"name": name, "home": home})

print("ƒê√£ th√™m sinh vi√™n v√†o file!")
```

### 7.4. T·∫°o GIF animation (V√≠ d·ª• t·ª´ b√†i gi·∫£ng)

```python
# V√≠ d·ª• 13: T·∫°o GIF ƒë∆°n gi·∫£n
from PIL import Image, ImageDraw

# T·∫°o 2 frames cho animation
frames = []

# Frame 1: H√¨nh tr√≤n ·ªü tr√°i
img1 = Image.new('RGB', (400, 200), color='white')
draw1 = ImageDraw.Draw(img1)
draw1.ellipse([50, 50, 150, 150], fill='blue')
frames.append(img1)

# Frame 2: H√¨nh tr√≤n ·ªü ph·∫£i
img2 = Image.new('RGB', (400, 200), color='white')
draw2 = ImageDraw.Draw(img2)
draw2.ellipse([250, 50, 350, 150], fill='blue')
frames.append(img2)

# L∆∞u th√†nh GIF
frames[0].save(
    'animation.gif',
    save_all=True,
    append_images=[frames[1]],
    duration=500,  # 500ms m·ªói frame
    loop=0  # L·∫∑p v√¥ h·∫°n
)

print("ƒê√£ t·∫°o animation.gif")
```
