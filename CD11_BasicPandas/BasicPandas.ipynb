## Các thành phần chính của `pandas`: `Series` và `DataFrames`

Hai thành phần chính của pandas là `Series` và `DataFrame`.

`Series` về cơ bản là một cột, và `DataFrame` là một bảng đa chiều được tạo thành từ một tập hợp các `Series`. 

<img src="assets/series-and-dataframe.png" width=600px />

`DataFrames` và `Series` khá giống nhau ở chỗ nhiều thao tác bạn có thể thực hiện với cái này thì bạn cũng có thể thực hiện với cái kia, như điền vào các giá trị null và tính trung bình.

Bạn sẽ thấy cách những thành phần này hoạt động khi chúng ta bắt đầu làm việc với dữ liệu ở phần dưới. 

### Tạo `DataFrames` từ đầu
Việc tạo `DataFrames` ngay trong `Python` là điều cần biết và rất hữu ích khi kiểm tra các phương pháp và chức năng mới bạn tìm thấy trong tài liệu `pandas`.

Có *nhiều* cách để tạo một DataFrame từ đầu, nhưng một lựa chọn tốt là chỉ sử dụng một `dict` đơn giản.

Giả sử chúng ta có một quầy bán trái cây bán táo và cam. Chúng ta muốn có một cột cho mỗi loại trái cây và một hàng cho mỗi lần mua hàng của khách hàng. Để sắp xếp điều này như một từ điển cho `pandas`, chúng ta có thể làm như sau:

```python
data = {
    'apples': [3, 2, 0, 1], 
    'oranges': [0, 3, 7, 2]
}
```

Và sau đó truyền nó vào hàm khởi tạo DataFrame của pandas:

```python
purchases = pd.DataFrame(data)

purchases
```

**Nó hoạt động như thế nào?**

Mỗi cặp *(key, value)* trong `data` tương ứng với một *cột* trong `DataFrame` kết quả.

**Index** của `DataFrame` này đã được cung cấp cho chúng ta khi tạo dưới dạng số từ 0-3, nhưng chúng ta cũng có thể tạo riêng `index` của mình khi khởi tạo `DataFrame`.

Hãy để tên khách hàng là `index` của chúng ta:

```python
import pandas as pd
data = {
    'apples': [3, 2, 0, 1], 
    'oranges': [0, 3, 7, 2]
}
purchases = pd.DataFrame(data, index=['June', 'Robert', 'Lily', 'David'])

purchases
purchases.loc['June']
```

Vì vậy, bây giờ chúng ta có thể định vị/**loc**ate đơn hàng của một khách hàng bằng cách sử dụng tên của họ:

```python
purchases.loc['June']
```

Sẽ có thêm thông tin về việc định vị và trích xuất dữ liệu từ `DataFrame` sau, nhưng bây giờ bạn nên tạo một `DataFrame` với bất kỳ dữ liệu ngẫu nhiên nào để học.

Hãy chuyển sang một số phương pháp nhanh chóng để tạo `DataFrames` từ các nguồn khác nhau.

Việc xem nhanh kiểu dữ liệu thực sự rất hữu ích. Hãy tưởng tượng bạn vừa nhập một số `JSON` và các số nguyên được ghi như là chuỗi. Bạn thực hiện một số phép toán và tìm thấy một ngoại lệ `"unsupported operand"` bởi vì bạn không thể làm toán học với chuỗi. Gọi `.info()` sẽ nhanh chóng chỉ ra rằng cột bạn nghĩ là tất cả các số nguyên thực ra là đối tượng chuỗi.

Một thuộc tính nhanh chóng và hữu ích khác là `.shape`, chỉ xuất ra một `tuple` của (số hàng, số cột):

```python
movies_df.shape
```

Lưu ý rằng `.shape` không có dấu ngoặc đơn và là một `tuple` đơn giản với định dạng (số hàng, số cột). Vì vậy, chúng tôi có **1000 hàng** và **11 cột** trong `DataFrame` phim của chúng tôi.

Bạn sẽ sử dụng `.shape` rất nhiều khi làm sạch và biến đổi dữ liệu. Ví dụ, bạn có thể lọc một số hàng dựa trên một số tiêu chí và sau đó muốn biết nhanh chóng bao nhiêu hàng đã bị loại bỏ.

### Xử lý dữ liệu trùng lặp

Tập dữ liệu này không có các dòng trùng lặp, nhưng luôn quan trọng khi xác minh bạn không đang tổng hợp các dòng trùng lặp.

Để minh họa, chúng ta chỉ cần nhân đôi `DataFrame` phim của chúng ta bằng cách nối nó vào chính nó:

```python
temp_df = movies_df._append(movies_df)

temp_df.shape
```

Sử dụng `_append()` sẽ trả về một bản sao mà không ảnh hưởng đến `DataFrame` gốc. Chúng tôi đang ghi lại bản sao này trong `temp_df` để chúng tôi không làm việc với dữ liệu thực.

Lưu ý gọi `.shape` nhanh chóng chứng minh số hàng của `DataFrame` của chúng tôi đã tăng gấp đôi.

Bây giờ chúng ta có thể thử loại bỏ các bản sao:

```python
temp_df = temp_df.drop_duplicates()

temp_df.shape
```

Giống như `_append()`, phương thức `drop_duplicates()` cũng sẽ trả về một bản sao của `DataFrame` của bạn, nhưng lần này với các bản sao đã bị loại bỏ. Gọi `.shape` xác nhận chúng tôi đã quay trở lại 1000 dòng của tập dữ liệu gốc của chúng tôi.

Việc liên tục gán `DataFrame` cho cùng một biến như trong ví dụ này hơi rườm rà. Vì lý do này, pandas có tham số từ khóa `inplace` trên nhiều phương thức của nó. Sử dụng `inplace=True` sẽ sửa đổi đối tượng `DataFrame` tại chỗ/trực tiếp:

```python
temp_df.drop_duplicates(inplace=True)
```

Bây giờ `temp_df` của chúng tôi *sẽ* có dữ liệu đã biến đổi tự động.

Tham số quan trọng khác cho `drop_duplicates()` là `keep`, có ba tùy chọn có thể:

* `first`: (mặc định) Loại bỏ các bản sao ngoại trừ lần xuất hiện đầu tiên.
* `last`: Loại bỏ các bản sao ngoại trừ lần xuất hiện cuối cùng.
* `False`: Loại bỏ tất cả các bản sao.

Vì chúng tôi không xác định tham số `keep` trong ví dụ trước nên nó được mặc định thành `first`. Điều này có nghĩa là nếu hai hàng giống nhau thì `pandas` sẽ loại bỏ hàng thứ hai và giữ hàng đầu tiên. Sử dụng `last` có hiệu ứng ngược lại: hàng đầu tiên sẽ bị loại bỏ.

`keep`, mặt khác (với tùy chọn `False`), sẽ loại bỏ tất cả các bản sao. Nếu hai hàng giống nhau thì cả hai sẽ bị loại bỏ. Xem điều gì xảy ra với `temp_df`:

```python
temp_df = movies_df._append(movies_df)  # make a new copy

temp_df.drop_duplicates(inplace=True, keep=False)

temp_df.shape
```

Vì tất cả các hàng đều là trùng lặp, `keep=False` đã loại bỏ tất cả chúng dẫn đến không còn hàng nào còn lại. Nếu bạn đang tự hỏi tại sao bạn muốn làm điều này, một lý do là nó cho phép bạn xác định tất cả các bản sao trong tập dữ liệu của bạn. Khi các lựa chọn có điều kiện được hiển thị bên dưới, bạn sẽ thấy cách làm điều đó.

### Làm thế nào để làm việc với giá trị bị thiếu

Khi khám phá dữ liệu, bạn rất có thể sẽ gặp các giá trị thiếu hoặc `null`, đó chính là các giữ chỗ cho các giá trị không tồn tại. Phổ biến nhất, bạn sẽ thấy `None` của Python hoặc `np.nan` của `NumPy`, mỗi loại đều được xử lý khác nhau trong một số tình huống.

Có hai lựa chọn khi xử lý giá trị `null`:

1. Loại bỏ các hàng hoặc cột với giá trị `null`
2. Thay thế giá trị `null` bằng giá trị không phải `null`, một kỹ thuật được biết đến là **imputation**/sự thay đổi

Hãy tính tổng số lượng giá trị `null` trong mỗi cột của tập dữ liệu của chúng tôi. Bước đầu tiên là kiểm tra xem các ô nào trong DataFrame của chúng tôi là `null`:

```python
movies_df.isnull()
```

Lưu ý rằng `isnull()` trả về một DataFrame nơi mỗi ô là `True` hoặc `False` tùy thuộc vào trạng thái `null` của ô đó.

Để đếm số lượng giá trị `null` trong mỗi cột, chúng tôi sử dụng một hàm tổng hợp để tính tổng:

```python
movies_df.isnull().sum()
```

`.isnull()` chỉ sử dụng riêng rẽ không phải là rất hữu ích, và thường được sử dụng cùng với các phương thức khác, như `sum()`.

Bây giờ chúng ta có thể thấy rằng dữ liệu của chúng ta có **128** giá trị thiếu cho cột `revenue_millions` và **64** giá trị thiếu cho cột `metascore`.

#### Loại bỏ giá trị `null`

Các Nhà khoa học dữ liệu và Phân tích viên thường xuyên đối mặt với thách thức về việc loại bỏ hoặc điền giá trị `null`, và đây là một quyết định đòi hỏi kiến thức sâu rộng về dữ liệu và ngữ cảnh của nó. Nói chung, việc loại bỏ dữ liệu `null` chỉ được đề xuất nếu bạn có một lượng nhỏ dữ liệu bị thiếu.

Loại bỏ giá trị `null` khá đơn giản:

```python
movies_df.dropna()
```

Thao tác này sẽ xóa bất kỳ **hàng** nào có ít nhất một giá trị `null`, nhưng nó sẽ trả về một `DataFrame` mới mà không thay đổi `DataFrame` gốc. Bạn cũng có thể chỉ định `inplace=True` trong phương thức này.

Vì vậy, trong trường hợp của tập dữ liệu của chúng tôi, thao tác này sẽ xóa 128 hàng nơi `revenue_millions` là `null` và 64 hàng nơi `metascore` là `null`. Rõ ràng điều này dường như là lãng phí vì có dữ liệu hoàn toàn tốt trong các cột khác của những hàng bị xóa. Đó là lý do tại sao chúng tôi sẽ xem xét imputation/sự thay đổi tiếp theo.

Không chỉ xóa các hàng, bạn cũng có thể xóa các cột với giá trị `null` bằng cách đặt `axis=1`:

### Imputation/Sự thay đổi

`Imputation` là một kỹ thuật trích chọn đặc trưng thông thường được sử dụng để giữ lại dữ liệu quý giá có giá trị `null`.

Có thể có trường hợp mà việc xóa mỗi hàng có giá trị `null` loại bỏ quá nhiều phần từ tập dữ liệu của bạn, vì vậy thay vào đó, chúng ta có thể điền giá trị `nul`l đó bằng một giá trị khác, thường là **mean** hoặc **median** của cột đó.

Hãy xem xét việc điền các giá trị thiếu trong cột `revenue_millions`. Đầu tiên, chúng ta sẽ trích xuất cột đó thành một biến riêng:

```python
revenue = movies_df['revenue_millions']
revenue_mean = revenue.mean()

revenue_mean
```

Sử dụng dấu ngoặc vuông là cách chung chúng ta chọn các cột trong `DataFrame`.

Nếu bạn nhớ lại khi chúng ta tạo `DataFrame` từ đầu, các khóa/key của `dict` đã trở thành tên cột. Bây giờ khi chúng ta chọn các cột của `DataFrame`, chúng ta sử dụng dấu ngoặc vuông giống như thể chúng ta đang truy cập vào một từ điển `Python`.

`revenue` bây giờ chứa một `Series`:

```python
print(movies_df)
revenue = movies_df['revenue_millions']
revenue.head()
```

```python
revenue_mean = revenue.mean()

revenue_mean
```

Điền toàn bộ cột bằng cùng một giá trị như thế này là một ví dụ cơ bản. Sẽ là một ý tưởng tốt hơn nếu thử điền một cách chi tiết hơn theo `Thể loại` hoặc `Đạo diễn`.

Ví dụ, bạn sẽ tìm ra giá trị trung bình của doanh thu được tạo ra trong từng thể loại riêng lẻ và điền các giá trị `null` trong từng thể loại bằng giá trị trung bình của thể loại đó.

### Hiểu các biến của bạn

Sử dụng `describe()` trên toàn bộ `DataFrame`, chúng ta có thể nhận được một bản tóm tắt về phân phối của các biến liên tục:

### Cắt, chọn, trích xuất `DataFrame`

Cho đến nay, chúng ta đã tập trung vào một số tổng kết cơ bản về dữ liệu của chúng ta. Chúng ta đã tìm hiểu về việc trích xuất cột đơn giản bằng cách sử dụng dấu ngoặc đơn, và chúng ta đã điền vào các giá trị `null` trong một cột bằng cách sử dụng `fillna()`. Dưới đây là các phương pháp cắt, chọn và trích xuất khác mà bạn sẽ cần sử dụng liên tục.

Điều quan trọng cần lưu ý là, mặc dù nhiều phương pháp giống nhau, nhưng `DataFrame` và `Series` có các thuộc tính khác nhau, vì vậy bạn cần chắc chắn biết bạn đang làm việc với loại nào, nếu không bạn sẽ nhận được lỗi thuộc tính.

Hãy xem xét việc làm việc với các cột đầu tiên.

#### Theo cột

Bạn đã thấy cách trích xuất một cột bằng cách sử dụng dấu ngoặc vuông như sau:

```python
genre_col = movies_df['genre']

type(genre_col)
```

Điều này sẽ trả về một `Series`. Để trích xuất một cột dưới dạng `DataFrame`, bạn cần truyền vào một danh sách các tên cột. Trong trường hợp của chúng ta, đó chỉ là một cột duy nhất:

```python
genre_col = movies_df[['genre']]

type(genre_col)
```

Vì chỉ là một danh sách, nên việc thêm tên cột khác rất dễ dàng:

```python
subset = movies_df[['genre', 'rating']]

subset.head()
```

Bây giờ chúng ta sẽ xem xét việc lấy dữ liệu theo hàng.

#### Theo hàng

Đối với hàng, chúng ta có hai lựa chọn:

- `.loc` - định vị theo tên
- `.iloc`- định vị theo chỉ số số học

Hãy nhớ rằng chúng ta vẫn được lập chỉ mục theo tiêu đề/Title phim, vì vậy để sử dụng `.loc` chúng ta cung cấp tiêu đề của một bộ phim:

```python
prom = movies_df.loc["Prometheus"]

prom
```

Mặt khác, với `iloc`, chúng ta cung cấp chỉ số số học của Prometheus:

```python
prom = movies_df.iloc[1]
prom
```

`loc` và `iloc` có thể được coi như tương tự như việc cắt `list` (list sclicing) trong Python. Để minh họa thêm, hãy chọn nhiều hàng.

Bạn sẽ làm như thế nào với một danh sách? Trong Python, chỉ cần cắt bằng dấu ngoặc vuông như `example_list[1:4]`. Việc này cũng hoạt động theo cùng cách trong `pandas`:

```python
movie_subset = movies_df.loc['Prometheus':'Sing']

movie_subset
```

```python
movie_subset = movies_df.iloc[1:4]

movie_subset
```

Một sự khác biệt quan trọng giữa việc sử dụng `.loc` và `.iloc` để chọn nhiều hàng là `.loc` bao gồm bộ phim *Sing* trong kết quả, nhưng khi sử dụng `.iloc` chúng ta nhận được các hàng từ `1` đến `4` nhưng bộ phim tại chỉ số `4` (*Suicide Squad*) không được bao gồm.

Cắt với `.iloc` tuân theo cùng một quy tắc như cắt với danh sách, đối tượng tại chỉ mục ở cuối không được bao gồm.

#### Lựa chọn theo điều kiện
Chúng tôi đã trình bày cách chọn các cột và hàng, nhưng nếu chúng tôi muốn thực hiện một lựa chọn theo điều kiện thì sao?

Ví dụ, nếu chúng tôi muốn lọc `DataFrame` phim của chúng tôi để chỉ hiển thị các phim do "Ridley Scott" đạo diễn hoặc các phim có xếp hạng lớn hơn hoặc bằng `8.0`?

Để làm điều đó, chúng ta lấy một cột từ `DataFrame` và áp dụng một điều kiện `Boolean` cho nó. Dưới đây là một ví dụ về điều kiện `Boolean`:

```python
condition = (movies_df['director'] == "Ridley Scott")

condition.head()
```

Tương tự như `isnull()`, điều này trả về một Series của các giá trị `True` và `False`: `True` cho các phim do "Ridley Scott" đạo diễn và `False` cho những phim không do ông đạo diễn.

Chúng ta muốn lọc ra tất cả các bộ phim không do "Ridley Scott" đạo diễn, nói cách khác, chúng ta không muốn các bộ phim `False`. Để trả về các hàng mà điều kiện đó là `True`, chúng ta phải đưa thao tác này vào `DataFrame`:

```python
movies_df[movies_df['director'] == "Ridley Scott"].head()
```

Bạn có thể làm quen với việc nhìn vào các điều kiện này bằng cách đọc nó như sau:

> `Chọn movies_df nơi movies_df director bằng Ridley Scott`

Hãy xem xét việc lựa chọn theo điều kiện sử dụng các giá trị số bằng cách lọc `DataFrame` theo xếp hạng:

```python
movies_df[movies_df['rating'] >= 8.6].head(3)
```

Chúng ta có thể tạo ra các điều kiện phong phú hơn bằng cách sử dụng các toán tử logic `|` cho "hoặc" và `&` cho "và".

Hãy lọc `DataFrame` để chỉ hiển thị các phim của "Christopher Nolan" HOẶC "Ridley Scott":

```python
movies_df[(movies_df['director'] == 'Christopher Nolan') | (movies_df['director'] == 'Ridley Scott')].head()
```

Chúng ta cần đảm bảo rằng nhóm các phép đánh giá bằng dấu ngoặc để `Python` biết cách đánh giá điều kiện.

Tuy nhiên, bằng cách sử dụng phương thức `isin()`, chúng ta có thể làm cho điều này ngắn gọn hơn:

```python
movies_df[movies_df['director'].isin(['Christopher Nolan', 'Ridley Scott'])].head()
```

Giả sử chúng ta muốn tất cả các bộ phim được phát hành giữa năm `2005` và `2010`, có xếp hạng trên `8.0`, nhưng thu nhập dưới phân vị thứ 25 (tứ phân vị thứ nhất - **Q1**).

Dưới đây là cách chúng ta có thể thực hiện tất cả những điều đó:

```python
movies_df[
    ((movies_df['year'] >= 2005) & (movies_df['year'] <= 2010))
    & (movies_df['rating'] > 8.0)
    & (movies_df['revenue_millions'] < movies_df['revenue_millions'].quantile(0.25))
]
```

Nếu bạn nhớ lại khi chúng tôi sử dụng `.describe()`, phân vị thứ 25 cho doanh thu là khoảng 17.4, và chúng tôi có thể truy cập trực tiếp vào giá trị này bằng cách sử dụng phương thức `quantile()` với một số thực là 0.25.

Vì vậy, ở đây chúng tôi chỉ có bốn bộ phim phù hợp với tiêu chí đó.

## Áp dụng các hàm

Có thể lặp qua `DataFrame` hoặc `Series` như bạn làm với một danh sách, nhưng việc làm như vậy - đặc biệt là trên các tập dữ liệu lớn - rất chậm.

Một lựa chọn hiệu quả hơn là `apply()` một hàm vào tập dữ liệu. Ví dụ, chúng ta có thể sử dụng một hàm để chuyển đổi các bộ phim có `8.0` hoặc cao hơn thành giá trị chuỗi "good" và phần còn lại thành "bad" và sử dụng các giá trị đã biến đổi này để tạo một cột mới.

Đầu tiên, chúng ta sẽ tạo một hàm mà khi được cung cấp một xếp hạng, xác định nó là "good" hay "bad":

```python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

Bây giờ chúng ta muốn gửi toàn bộ cột xếp hạng qua hàm này, đó là những gì `apply()` làm:

```python
movies_df["rating_category"] = movies_df["rating"].apply(rating_function)

movies_df.head(2)
```

Phương thức `.apply()` chuyển mọi giá trị trong cột `rating` qua `rating_function` và sau đó trả về một `Series` mới. `Series` này sau đó được gán cho một cột mới có tên `rating_category`.

Bạn cũng có thể sử dụng các hàm ẩn danh cũng được. Hàm `lambda` này đạt được cùng một kết quả như `rating_function`:

```python
movies_df["rating_category"] = movies_df["rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

movies_df.head(2)
```

Nói chung, việc sử dụng `apply()` sẽ nhanh hơn nhiều so với việc lặp qua các dòng một cách thủ công bởi vì `pandas` đang sử dụng `vectorization`.

> `Vectorization`: một phong cách lập trình máy tính nơi các phép toán được áp dụng cho cả các mảng thay vì các phần tử cá nhân —[Wikipedia](https://en.wikipedia.org/wiki/Vectorization)

Một ví dụ tốt về việc sử dụng nâng cao `apply()` là trong công việc xử lý ngôn ngữ tự nhiên (`NLP`). Bạn sẽ cần áp dụng tất cả các loại hàm làm sạch văn bản cho các chuỗi để chuẩn bị cho học máy.

## Vẽ đồ thị ngắn gọn

Điều tuyệt vời khác về `pandas` là nó tích hợp với `Matplotlib`, vì vậy bạn có khả năng vẽ đồ thị trực tiếp từ `DataFrames` và `Series`. Để bắt đầu, chúng ta cần nhập `Matplotlib` (`pip install matplotlib`):

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)}) # set font and plot size to be larger
```

Bây giờ chúng ta có thể bắt đầu vẽ. Sẽ không có nhiều phần bao phủ về việc vẽ đồ thị, nhưng nó nên đủ để khám phá dữ liệu của bạn một cách dễ dàng.

**Ghi chú:**
Đối với các **biến phân loại**, hãy sử dụng *Biểu đồ cột* và *Boxplots*. Đối với các **biến liên tục**, hãy sử dụng *Histograms*, *Scatterplots*, *đồ thị đường/Line* và *Boxplots*.

Hãy vẽ biểu đồ mối quan hệ giữa xếp hạng và doanh thu. Tất cả những gì chúng ta cần làm là gọi `.plot()` trên `movies_df` với một số thông tin về cách xây dựng biểu đồ:

Vấn đề gì với dấu chấm phẩy/`;` (sau hàm plot())?  Đó không phải là lỗi cú pháp, chỉ là một cách để ẩn output `<matplotlib.axes._subplots.AxesSubplot at 0x26613b5cc18>` khi vẽ đồ thị trong Jupyter notebooks.
